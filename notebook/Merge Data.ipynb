```python
import pandas as pd
import numpy as np
import glob
```

```python
path = r'/content/drive/MyDrive/Scrapping Web Tripadvisor/cleaned_data' 
all_files = glob.glob(path + "/*.csv")
print(all_files)
```

```python
li = []
for f in all_files:
  place = f[62:-4]
  df = pd.read_csv(f, index_col=None, header=0)
  df.insert(loc=0, column='place', value=place)
  df = df.drop(columns=['rating', 'date'])
  li.append(df)

df_all = pd.concat(li, axis=0, ignore_index=True)
```
